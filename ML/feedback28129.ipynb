```python
import pandas as pd
import numpy as np

data1 = pd.read_excel('C:/Users/Krishu Thapa/Desktop/data.xls')
data1 = data1.replace(
{
    "What was your SLC percentage?":'[A-Za-z()%]',
    "What was your +2 percentage?":'[A-Za-z()%]',
    "What was your KUCAT score?":'[A-Za-z()%]',
    "Internal I percentage" :'[A-Za-z()%]',
    "Internal II percentage" :'[A-Za-z()%]',
    "How is your class attendance ?" :'[A-Za-z()%]'
},'' ,regex=True
)

df1 = data1['What was your KUCAT score?']
x=np.array(df1)
y=x.astype(int)

import pandas as pd
data2 = pd.DataFrame(data1)
subset=data2.iloc[:,:]
subset["What was your KUCAT score?"]=y

data2=pd.get_dummies( data2,columns=['Have you cleared all your papers you have appeared in?',
                                      'How seriously do you take your pre examination breaks?',
                                      'What is your hours of self study per day?',
                                    'Do you do your assignment yourself? '])
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

b=data1['Your GPA till now ?']

X, y = data2, b
print(X.shape)
X_new = SelectKBest(f_regression, k=12)
X_new.fit(X,y)
X_new1 = X_new.transform(X)  #transform(X, y)
print (X_new1.shape)
```

```python
from sklearn.cross_validation import train_test_split
a=data1['Your GPA till now ?']
X_train, X_test,y_train, y_test=train_test_split(X_new1,a,random_state=1)

print (X_test.shape)
print (X_train.shape)
print (y_train.shape)
print (y_test.shape)

```

```python
from sklearn import linear_model

linreg=linear_model.LinearRegression()
linreg.fit(X_train,y_train)
```

```python
print(linreg.intercept_)
print(linreg.coef_)

```

```python
print(y_test)
```

```python
y_pred=linreg.predict(X_test)
```
